### Cleaning Data

Missing value on Product Data

```python
products_df.isna().sum()
```

```python
products_df.product_category_name.fillna(value="Not Identified", inplace=True)
```

```python
products_df.product_category_name.value_counts()
```

```python
products_df.product_name_lenght.fillna(value=0, inplace=True)
products_df.product_description_lenght.fillna(value=0, inplace=True)
products_df.product_photos_qty.fillna(value=0, inplace=True)
products_df.product_weight_g.fillna(value=0, inplace=True)
products_df.product_length_cm.fillna(value=0, inplace=True)
products_df.product_height_cm.fillna(value=0, inplace=True)
products_df.product_width_cm.fillna(value=0, inplace=True)
```

```python
products_df.info()
```

Duplicated on Location Data

```python
location_df.drop_duplicates(inplace=True)
```

```python
location_df.duplicated().sum()
```

Unsuitable Data on Orders Desc Data

```python
datetime_columns = ["order_delivered_customer_date", "order_approved_at", "order_delivered_carrier_date", "order_estimated_delivery_date", "order_purchase_timestamp"]
for column in datetime_columns:
  order_desc_df[column] = pd.to_datetime(order_desc_df[column])
```

## Exploratory Data Analysis (EDA)

### Explore Products and Orders Data

```python
orders_df = pd.merge(
    left=orders_df,
    right=products_df,
    how="left",
    left_on="product_id",
    right_on="product_id"
)
orders_df.head()
```

```python
orders_df.info()
```

```python
orders_df = pd.merge(
    left=orders_df,
    right=products_eng_df,
    how="left",
    left_on="product_category_name",
    right_on="product_category_name"
)
orders_df.head()
```

```python
orders_df.groupby(by="product_category_name_english").agg({
    "product_id": "nunique",
    "order_item_id": "count",
    "price": ["min", "max", "sum"]
})
```

### Explore Orders, Customers and Locations Data

```python
order_desc_df = pd.merge(
    left=orders_df,
    right=order_desc_df,
    how="left",
    left_on="order_id",
    right_on="order_id"
)
order_desc_df.head()
```

```python
cust_df = pd.merge(
    left=cust_df,
    right=location_df,
    how="left",
    left_on="customer_zip_code_prefix",
    right_on="geolocation_zip_code_prefix"
)
cust_df.head()
```

```python
cust_desc_df = pd.merge(
    left=cust_df,
    right=order_desc_df,
    how="left",
    left_on="customer_id",
    right_on="customer_id"
)
cust_desc_df.head()
```

```python
cust_desc_df.info()
```

```python
cust_desc_df.groupby(by="customer_city").customer_id.nunique().sort_values(ascending=False).head(10)
```

```python
cust_desc_df.groupby(by=["customer_city", "product_category_name_english"]).agg({
    "order_item_id": "count",
    "price": "sum"
}).head(5)
```

```python
cust_desc_df.groupby(by="order_status").agg({
    "product_category_name_english": "nunique",
    "customer_id": "nunique",
    "customer_city": "nunique",
    "customer_state": "nunique",
})
```

```python
delivery_time = cust_desc_df["order_delivered_customer_date"] - cust_desc_df["order_purchase_timestamp"]
delivery_time = delivery_time.apply(lambda x: x.total_seconds())
cust_desc_df["delivery_time"] = round(delivery_time/86400)
```

```python
cust_desc_df.sample(5)
```

```python
cust_desc_df.groupby(by="delivery_time").agg({
    "product_id": "nunique",
    "order_item_id": "count",
}).head(10)
```

## Visualization & Explanatory Analysis

### Which products yield the highest sales and make the most substantial contribution to the company's revenue?

```python
revenue_order_items_df = orders_df.groupby("product_category_name_english").price.sum().sort_values(ascending=False).reset_index()
revenue_order_items_df.head(15)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

color1 = ["#228B22", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
color2 = ["#D2042D", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="price", y="product_category_name_english", data=revenue_order_items_df.head(5), palette=color1, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Performing Product", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)

sns.barplot(x="price", y="product_category_name_english", data=revenue_order_items_df.sort_values(by="price", ascending=True).head(5), palette=color2, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Worst Performing Product", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)

plt.suptitle("Product Contribution to Company Revenue", fontsize=20)
plt.show()
```

```python
total_order_items_df = orders_df.groupby("product_category_name_english").order_item_id.count().sort_values(ascending=False).reset_index()
total_order_items_df.head(15)
```

```python
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))

color1 = ["#228B22", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]
color2 = ["#D2042D", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="order_item_id", y="product_category_name_english", data=total_order_items_df.head(5), palette=color1, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Best Performing Product", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)

sns.barplot(x="order_item_id", y="product_category_name_english", data=total_order_items_df.sort_values(by="order_item_id", ascending=True).head(5), palette=color2, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Worst Performing Product", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)

plt.suptitle("Product Contribution to Company Revenue", fontsize=20)
plt.show()
```
